```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
austin = "CSV/Austin_Counts.csv"
louisville = "CSV/Louisville_Counts.csv"
austin_df = pd.read_csv(austin, encoding = 'latin-1')
louisville_df = pd.read_csv(louisville, encoding = 'latin-1')
```

```python
austin_df
```

```python
louisville_df
```

```python
#data frame for grouped Bar Chart to compare Adoption rates
austin_per = austin_df['% of Animals Adopted']
louisville_per = louisville_df['% of Animals Adopted']
austin_eu_per = austin_df['% of Animals Euthenized']
louisville_eu_per = louisville_df['% of Animals Euthenized']
age = austin_df['Age Range']
per_df =pd.DataFrame({'Age Range': age,
     'Austin % of Dog Adoptions 2013-2017': austin_per,
     'Austin % of Dog Euthanizations 2013-2017': austin_eu_per,
     'Louisville % of Dog Adoptions 2013-2017': louisville_per,
     'Louisviell % of Dog Euthanizations 20123-2017': louisville_eu_per
    })
per_df
```

```python
#plot goruped bar
#grouped and stacked bar chart to visually show differences in Cat Colors
aus_per = np.array(austin_per)
aus_eu_per = np.array(austin_eu_per)
lou_per= np.array(louisville_per)
lou_eu_per = np.array(louisville_eu_per)

age_list =per_df['Age Range']

#create subplots
fig, ax = plt.subplots(figsize=(20,10))
plt.style.use('seaborn')

#plot parameters
pos = list(range(len(age_list))) 
width = 0.4

#["lightsteelblue","cornflowerblue","royalblue","mediumblue","darkblue","slateblue","blueviolet"]
#catColors = ["khaki","gold","goldenrod","darkgoldenrod","darkkhaki","olive","darkolive"]

aus1c_plot = plt.bar(pos, aus_per, width, alpha=0.5, color='royalblue')
aus2c_plot = plt.bar(pos, aus_eu_per, width, alpha=0.5, bottom=aus_per, color='cornflowerblue')
lou1c_plot = plt.bar([p + width for p in pos], lou_per, width, alpha=0.5, color='gold')
lou2c_plot = plt.bar([p + width for p in pos], lou_eu_per, width, alpha=0.5, bottom=lou_per, color='goldenrod')

# Set the y axis label
ax.set_ylabel('% of Dogs')
ax.set_xlabel('Age Group')

# Set the chart's title
ax.set_title('% of Dog Adoptions vs Euthanizations for per Age Groups (Austin, TX vs Louisville, KY)')


# Set the position of the x ticks
ax.set_xticks([p + .5 * width for p in pos])

# Set the labels for the x ticks
ax.set_xticklabels(age_list)

# Setting the x-axis and y-axis limits
#plt.xlim(min(pos)-width, max(pos)+width*4)
#plt.ylim([0, max(df['pre_score'] + df['mid_score'] + df['post_score'])] )

# Adding the legend and showing the plot
plt.legend(['Austin Dog Adoptions % 2013-2017', 'Austin Dog Euthaniztions % 2013-2017', 'Louisville Dog Adoptions2013-2017', 'Louisville Dog Euthanizations2013-2017'], loc='upper right')
plt.savefig("figs/age_dog.png")


plt.show()

```

```python
breed1 = "CSV/Austin_Counts_Breed.csv"
breed2 = "CSV/Louisville_Counts_Breed.csv"
austin_breed_df = pd.read_csv(breed1, encoding = 'latin-1')
louisville_breed_df = pd.read_csv(breed2, encoding = 'latin-1')
```

```python
austin_breed = austin_breed_df['Adoptions']
austin_eu_breed = austin_breed_df['Euthanizations']
austin_total_breed = austin_breed_df['Total Dogs'] 
louisville_breed = louisville_breed_df['Adoptions']
louisville_eu_breed = louisville_breed_df['Euthanizations']
louisville_total_breed =louisville_breed_df['Total Dogs'] 
breeds = austin_breed_df['Dog Breed']
totals_breed_df =pd.DataFrame({'Dog Breed': breeds,
     'Austin Adoptions by Dog Breed 2013-2017': austin_breed,
     'Austin Euthanizations by Dog Breed 2013-2017': austin_eu_breed,
     'Austin Total Dogs by Breed 2013-2017': austin_total_breed,
     'Louisville Adoptions by Dog Breed 2013-2017': louisville_breed,
     'Louisville Euthanizations by Dog Breed 2013-2017': louisville_eu_breed,
     'Louisville Total Dogs by Breed 2013-2017': louisville_total_breed
    })
totals_breed_df
```

```python
#grouped and stacked bar chart to visually show differences in Cat Colors
aus_breed = np.array(austin_breed)
aus_eu_breed = np.array(austin_eu_breed)
aus_total_breed = np.array(austin_total_breed)
lou_breed= np.array(louisville_breed)
lou_eu_breed = np.array(louisville_eu_breed)
lou_total_breed= np.array(louisville_total_breed)
age_list =list(str(breeds))

#create subplots
fig, ax = plt.subplots(figsize=(20,10))
plt.style.use('seaborn')

#plot parameters
pos = list(range(len(breeds))) 
width = 0.4

#["lightsteelblue","cornflowerblue","royalblue","mediumblue","darkblue","slateblue","blueviolet"]
#catColors = ["khaki","gold","goldenrod","darkgoldenrod","darkkhaki","olive","darkolive"]

aus1c_plot = plt.bar(pos, aus_total_breed, width, alpha=0.5, color='royalblue')
aus2c_plot = plt.bar(pos, aus_breed, width, alpha=0.5, bottom=aus_total_breed, color='cornflowerblue')
aus3c_plot = plt.bar(pos, aus_eu_breed, width, alpha=0.5, bottom=(aus_total_breed+aus_breed), color='lightsteelblue')
lou1c_plot = plt.bar([p + width for p in pos], lou_total_breed, width, alpha=0.5, color='darkgoldenrod')
lou2c_plot = plt.bar([p + width for p in pos], lou_breed, width, alpha=0.5, bottom=lou_total_breed, color='goldenrod')
lou2c_plot = plt.bar([p + width for p in pos], lou_eu_breed, width, alpha=0.5, bottom=(lou_total_breed+lou_breed), color='gold')

# Set the y axis label
ax.set_ylabel('Number of Dogs')
ax.set_xlabel('Dog Breeds')

# Set the chart's title
ax.set_title('Adoption/Euthanization Data for Popular Dog Breeds (Austin, TX vs Louisville, KY)')


# Set the position of the x ticks
ax.set_xticks([p + .5 * width for p in pos])

# Set the labels for the x ticks
ax.set_xticklabels(breeds)

# Setting the x-axis and y-axis limits
#plt.xlim(min(pos)-width, max(pos)+width*4)
#plt.ylim([0, max(df['pre_score'] + df['mid_score'] + df['post_score'])] )

# Adding the legend and showing the plot
plt.legend(['Austin Dogs 2013-2017', 'Austin Dog Adoptions 2013-2017', 'Austin Dog Euthanizations 2013-2017', 'Louisville Dog 2013-2017', 'Lousiville Dog Adoptions 2013-2017', 'Louisville Dog Euthanizations 2013-2017'], loc='upper right')
#plt.savefig("figs/age_adopt.png")


plt.show()

```
