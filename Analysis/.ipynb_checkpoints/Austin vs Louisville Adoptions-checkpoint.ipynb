```python
#data frame for grouped Bar Chart to compare Adoption rates
austin_per = austin_df['% of Animals Adopted']
louisville_per = louisville_df['% of Animals Adopted']
age = austin_df['Age Range']
per_df =pd.DataFrame({'Age Range': age,
     'Austin % of Dog Adoptions 2013-2017': austin_per,
     'Louisville % of Dog Adoptions 2013-2017': louisville_per,
    })
per_df
```

```python
#plot goruped bar
per_plot = per_df.plot.bar(x='Age Range', rot=0, title='Adoption Rates Austin, TX vs Louisville, KY', figsize=(15,10), fontsize=12)
per_plot.set_ylabel("% of Dogs Adopted")
plt.show()
```

```python
#data frame for grouped bar chart Euthanizations
austin_pere = austin_df['% of Animals Euthenized']
louisville_pere = louisville_df['% of Animals Euthenized']
pere_df =pd.DataFrame({'Age Range': age,
     'Austin % of Dog Euthenizations 2013-2017': austin_pere,
     'Louisville % of Dog Euthenizations 2013-2017': louisville_pere,
    })
pere_df
```

```python
#plot goruped bar
pere_plot = pere_df.plot.bar(x='Age Range', rot=0, title='Euthanizations Rates Austin, TX vs Louisville, KY', figsize=(15,10), fontsize=12)
pere_plot.set_ylabel("% of Dogs Euthanized")
plt.show()
```
