```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
austin = "CSV/Austin_Counts.csv"
louisville = "CSV/Louisville_Counts.csv"
austin_df = pd.read_csv(austin, encoding = 'latin-1')
louisville_df = pd.read_csv(louisville, encoding = 'latin-1')
```

```python
austin_df
```

```python
louisville_df
```

```python
austin_dog = austin_df['Adoptions']
austin_eu = austin_df['Euthanizations']
austin_total = austin_df['Total Dogs'] 
louisville_dog = louisville_df['Adoptions']
louisville_eu = louisville_df['Euthanizations']
louisville_total =louisville_df['Total Dogs'] 
age = austin_df['Age Range']
totals_df =pd.DataFrame({'Age Range': age,
     'Austin Dog Adoptions 2013-2017': austin_dog,
     'Austin Dog Euthanizations 2013-2017': austin_eu,
     'Austin Total Dogs 2013-2017': austin_total,
     'Louisville Dog Adoptions 2013-2017': louisville_dog,
     'Louisville Dog Euthanizations 203-2017': louisville_eu,
     'Louisville Total Dogs 2013-2017': louisville_total
    })
totals_df
```

```python
cat1 = "CSV/Austin_Counts_Cat.csv"
cat2 = "CSV/Louisville_Counts_Cat.csv"
austin_cat_df = pd.read_csv(cat1, encoding = 'latin-1')
louisville_cat_df = pd.read_csv(cat2, encoding = 'latin-1')
```

```python
#Cat DF for plotting
austin_cat = austin_cat_df['Adoptions']
austin_eu_cat = austin_cat_df['Euthanizations']
austin_total_cat = austin_cat_df['Total Cats'] 
louisville_cat = louisville_cat_df['Adoptions']
louisville_eu_cat = louisville_cat_df['Euthanizations']
louisville_total_cat =louisville_cat_df['Total Cats'] 
age_cat = austin_cat_df['Age Range']
totals_cat_df =pd.DataFrame({'Age Range': age_cat,
     'Austin Cat Adoptions 2013-2017': austin_cat,
     'Austin Cat Euthanizations 20123-2017': austin_eu_cat,
     'Austin Total Cats 2013-2017': austin_total_cat,
     'Louisville Cat Adoptions 2013-2017': louisville_cat,
     'Louisville Cat Euthanizations 2013-2017': louisville_eu_cat,
     'Louisville Total Cat 2013-2017': louisville_total_cat
    })
totals_cat_df
```

```python
animal_df = pd.merge(totals_df, totals_cat_df, on='Age Range', how='outer')
```

```python
animal_young_df = animal_df[animal_df["Age Range"].isin(["<1", "1-3"])]
age_young = animal_young_df["Age Range"]
age_young
animal_young_df
```

```python
animal_old_df = animal_df[animal_df["Age Range"].isin(["3-5", "5-7", "7-9", ">9"])]
animal_old_df
```
