```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#map csv
austin = "CSV/MergedData.csv"
#austin_df = pd.read_csv(austin)
austin_df = pd.read_csv(austin, encoding = 'latin-1')
austin_df.head()
```

```python
dog_df = austin_df.loc[(austin_df['Animal Type_intake'] == "Dog"), : ]
dog_df.head()
```

```python
dog_df["AgeYears_Outcome"] = round(dog_df["AgeDays_outcome"] / 365,2)
age_bins = [0,1,3,5,7,9, 25]
age_labels = ["<1", "1-3", "3-5", "5-7", "7-9", ">9"]
dog_df.head()

#Make Bins for Age
#hp_bins = [180, 200, 350, 400]
#hp_labels = ["Slow", "Decent", "Fast"]
```

```python
age_bins = [0,1,3,5,7,9, 25]
age_labels = ["<1", "1-3", "3-5", "5-7", "7-9", ">9"]
dog_df["Age Range"] = pd.cut(dog_df["AgeYears_Outcome"], age_bins, labels=age_labels)
dog_group = dog_df.groupby("Age Range")
dog_groups = dog_group["Outcome Type"].value_counts()
dog_groups_df = pd.DataFrame(dog_groups)
dog_groups_df.unstack(level = 0)
```
