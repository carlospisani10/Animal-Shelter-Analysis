## Common Distribution Types:
SciPy Reference: https://docs.scipy.org/doc/scipy/reference/stats.html

Cloudera Reference: https://blog.cloudera.com/blog/2015/12/common-probability-distributions-the-data-scientists-crib-sheet/


```python
# Dependencies
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
from matplotlib import style
style.use('ggplot')
```

## Normal Distribution

```python
# Generating a Random Distribution
normal_distribution = stats.gennorm.rvs(1, size=10000)
plt.scatter(np.arange(0, 10000), normal_distribution)
plt.show()
```

```python
# Visualize Data as Histogram
plt.hist(normal_distribution, 50, facecolor='green', alpha=0.75)
plt.show()
```

## Poisson Distribution

```python
# Generating a Random Distribution
poisson_distribution = stats.poisson.rvs(loc=10, mu=35, size=10000)
plt.scatter(np.arange(0, 10000), poisson_distribution)
plt.show()
```

```python
# Visualize as Histogram
plt.hist(poisson_distribution, 50, facecolor='green', alpha=0.75)
plt.show()
```

## Exponential Distribution

```python
# Generating a Random Distribution
exponential_distribution = stats.expon.rvs(size=10000)
plt.scatter(np.arange(0, 10000), exponential_distribution)
plt.show()
```

```python
# Visualize as Histogram
plt.hist(exponential_distribution, 50, facecolor='green', alpha=0.75)
plt.show()
```

# Normality Testing

```python
# Kolmogorov-Smornov test
# https://docs.scipy.org/doc/scipy-0.19.1/reference/generated/scipy.stats.kstest.html

# Generate some fake data
some_distribution = stats.gennorm.rvs(1, size=10000)

# Testing it for a normal distribution
stats.kstest(some_distribution, 'norm')
```
