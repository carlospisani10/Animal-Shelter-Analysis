```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
from matplotlib import style
style.use('ggplot')
```

```python
normal_distribution = stats.gennorm.rvs(1, size=1000)
plt.scatter(np.arange(0, 1000), normal_distribution)
plt.show()
```

```python
plt.hist(normal_distribution, 50, facecolor='green', alpha=0.75)
plt.show()
```
